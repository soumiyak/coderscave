```python
sales_by_store = df.groupby("STORECODE").sum().sort_values('BILL_AMT', ascending=False)
plt.figure(figsize=(15, 8))
sns.barplot(x=sales_by_store['BILL_AMT'], y=sales_by_store.index, data=sales_by_store, palette='viridis')
plt.title("Total Sales by Store")
plt.xlabel("Sales")
plt.ylabel("Store")
plt.grid(axis='x', color='black', linestyle='--', linewidth=0.5)
plt.xticks(rotation=0, ha='right')
plt.show()
```

```python
unique_storecodes = df['STORECODE'].unique()
plt.figure(figsize=(15, 20))

for i, storecode in enumerate(unique_storecodes[:10]):
    plt.subplot(5, 2, i+1)
    sns.scatterplot(x='STORECODE', y='BILL_AMT', data=df[df['STORECODE'] == storecode], alpha=0.6)
    plt.title(f"Scatter Plot - Store Code {storecode}")
    plt.xlabel("Store Code")
    plt.ylabel("Transaction Amount")

plt.tight_layout()
plt.show()
```

```python
top_10_stores = df.groupby('STORECODE')['BILL_AMT'].sum().nlargest(10).index

# Filter the DataFrame for the top 10 store codes
df_top_10_stores = df[df['STORECODE'].isin(top_10_stores)]

# Iterate over each month
for month in sorted(df_top_10_stores['MONTH'].unique()):
    # Filter data for the current month
    df_month = df_top_10_stores[df_top_10_stores['MONTH'] == month]

    # Create a bar chart for the current month
    plt.figure(figsize=(12, 6))
    sns.barplot(x='STORECODE', y='BILL_AMT', data=df_month, ci=None)
    plt.title(f"Sales - Month {month}")
    plt.xlabel("Store Code")
    plt.ylabel("Total Sales")

    # Show the chart
    plt.show()
```

```python
for i in df.STORECODE.unique():
    store_data = df.loc[df.STORECODE == i].groupby("DAY").sum()

    plt.figure(figsize=(20, 5))

    # Bar plot for total sales
    plt.subplot(1, 2, 1)
    sns.barplot(x=store_data.index, y=store_data["BILL_AMT"], data=store_data)
    plt.axhline(store_data["BILL_AMT"].mean(), color='purple')
    plt.title("Total Sales from Store: " + i)
    plt.xlabel("Day")
    plt.ylabel("Total Sales")
    plt.grid(axis='y', color='black')

    # Box plot for sales distribution
    plt.subplot(1, 2, 2)
    sns.boxplot(x=store_data["BILL_AMT"])
    plt.title("Sales Distribution - Store: " + i)
    plt.xlabel("Total Sales")

    # Adjust layout
    plt.tight_layout()

    # Show the charts
    plt.show()
```

```python
top_items_sales = df.groupby('GRP').sum().sort_values("VALUE",ascending=False)[['QTY','PRICE','VALUE']][0:25]
plt.figure(figsize=(12,8))
sns.barplot(x='VALUE',y=top_items_sales.index,data=top_items_sales)
plt.title("Sales from Top 25 Categories")
plt.xlabel("Sales")
plt.ylabel("Categories")
plt.grid(axis='x',color='black');
```

```python
for i in df.STORECODE.unique():
    x = df.loc[df.STORECODE == i].groupby("GRP").sum().sort_values("VALUE",ascending=False)[["VALUE","QTY"]][0:25]
    plt.figure(figsize=(12,8))
    sns.barplot(x='VALUE',y=x.index,data=x)
    plt.title("Top 25 Categories by Sales from store:  " + i)
    plt.ylabel("Categories")
    plt.xlabel("Sales")
```

```python
grp_sales = df.groupby('GRP')['BILL_AMT'].sum().sort_values(ascending=False)

# Take the top 10 GRPs
top_10_grp = grp_sales.head(10)

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_grp.index, y=top_10_grp.values)
plt.title('Top 10 GRPs Based on Sales')
plt.xlabel('GRP')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Show the chart
plt.show()
```

```python
grp_sales = df.groupby('GRP')['BILL_AMT'].sum().sort_values()

# Take the least 10 GRPs
least_10_grp = grp_sales.head(10)

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=least_10_grp.index, y=least_10_grp.values)
plt.title('Least 10 GRPs Based on Sales')
plt.xlabel('GRP')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Show the chart
plt.show()
```

```python
brd_value = df.groupby('BRD')['VALUE'].sum().sort_values(ascending=False)

# Take the top 10 BRDs
top_10_brd = brd_value.head(10)

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_brd.index, y=top_10_brd.values)
plt.title('Top 10 BRDs Based on Value')
plt.xlabel('BRD')
plt.ylabel('Total Value')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Show the chart
plt.show()
```

```python
grp_quantity = df.groupby('GRP')['QTY'].sum().sort_values(ascending=False)

# Take the top 10 GRPs
top_10_grp = grp_quantity.head(10)

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_grp.index, y=top_10_grp.values)
plt.title('Top 10 GRPs Based on Quantity')
plt.xlabel('GRP')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Show the chart
plt.show()
```
